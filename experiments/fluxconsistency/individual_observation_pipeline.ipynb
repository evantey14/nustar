```python
import config, os, sys, itertools
sys.path.append(os.path.join(os.getcwd(), os.pardir, os.pardir)) # add path to nutools. TODO: make more robust
import nutools
```

```python
# define data of interest (ie the values you want to iterate over)
observations = ['30001002001', '30001002003', '30001002004', '30302006004', '30302006006', '30302007002', '30302007004']
regions = ['NE', 'SW']
modules = ['A', 'B']
```

```python
# call nuproducts
for observation, module, region in list(itertools.product(observations, modules, regions)):
    indir = config.DATA_PATH + '/' + \
            observation + '/' + \
            'pipeline_clean' + '/'

    outdir = config.PRODUCT_PATH + '/'

    instrument = 'FPM' + module

    steminputs = 'nu' + observation

    stemout = 'nu' + observation + module + region

    srcregionfile = config.REGION_PATH + '/' + \
                    'Source_' + region + '.reg'

    bkgregionfile = config.REGION_PATH + '/' + \
                    'bg_' + observation + module + '.reg'

    evtfile = indir + steminputs + module + '01_cl.evt'
    
    cmd = nutools.nuproducts(indir=indir, 
                            outdir=outdir, 
                            instrument=instrument, 
                            steminputs=steminputs, 
                            stemout=stemout, 
                            srcregionfile=srcregionfile, 
                            bkgregionfile=bkgregionfile, 
                            evtfile=evtfile)
    !{cmd} # must call the command like this because of how nuproducts works
    # TODO: output useful information

# it's recommeneded to use fv or other tools to check the correctness of products
```

```python
for observation, module, region in list(itertools.product(observations, modules, regions)):
    stem = 'nu' + observation + module + region
    
    nutools.group_pha(infile = config.REL_PRODUCT_PATH + '/' + stem + '_sr.pha',
                      bgfile = config.REL_PRODUCT_PATH + '/' + stem + '_bk.pha',
                      outfile = config.REL_PRODUCT_PATH + '/' + stem + '_group.dat',
                      minbincount = '20',
                      is_individual = True)
    
    # needs rel product path because absolute path is too long
    nutools.bin_pha(infile = config.PRODUCT_PATH + '/' + stem + '_sr.pha',
                    grpfile = config.REL_PRODUCT_PATH + '/' + stem + '_group.dat',
                    outfile = config.REL_PRODUCT_PATH + '/' + stem + '_sr_bin.pha')
```
