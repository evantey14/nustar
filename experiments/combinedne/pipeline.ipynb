```python
import config, os, sys, itertools
sys.path.append(os.path.join(os.getcwd(), os.pardir, os.pardir)) # add path to nutools. TODO: make more robust
import nutools
```

```python
years = ['2012', '2017']
observationsbyyear = {'2012': ['30001002001', '30001002003', '30001002004'], 
                '2017': ['30302006004', '30302006006', '30302007002', '30302007004']}
observations = ['30001002001', '30001002003', '30001002004', '30302006004', '30302006006', '30302007002', '30302007004']
modules = ['A', 'B']
region = 'NE'
```

```python
# call nuproducts
for observation, module in list(itertools.product(observations, modules)):
    indir = config.DATA_PATH + '/' + \
            observation + '/' + \
            'pipeline_clean' + '/'

    outdir = config.PRODUCT_PATH + '/'

    instrument = 'FPM' + module

    steminputs = 'nu' + observation

    stemout = 'nu' + observation + module + region

    srcregionfile = config.REGION_PATH + '/' + \
                    'Source_' + region + '.reg'

    bkgregionfile = config.REGION_PATH + '/' + \
                    'bg_' + observation + module + '.reg'

    evtfile = indir + steminputs + module + '01_cl.evt'
    
    cmd = nutools.nuproducts(indir=indir, 
                            outdir=outdir, 
                            instrument=instrument, 
                            steminputs=steminputs, 
                            stemout=stemout, 
                            srcregionfile=srcregionfile, 
                            bkgregionfile=bkgregionfile, 
                            evtfile=evtfile)
    !{cmd}
```

```python
# combine yearly spectra
for year, module in list(itertools.product(years, modules)):
    input_obs = ['nu' + observation + module + region + '_sr.pha' 
                 for observation in observationsbyyear[year]]
    outfile = year + '_' + module + '_' + region
    nutools.addspec(config.EXPERIMENT_PATH, config.PRODUCT_PATH, input_obs, outfile)
```

```python
# group and bin yearly spectra
for year, module in list(itertools.product(years, modules)):
    stem = year + '_' + module + '_' + region
    
    nutools.group_pha(infile = config.REL_PRODUCT_PATH + '/' + stem + '.pha',
                      bgfile = config.REL_PRODUCT_PATH + '/' + stem + '.bak',
                      outfile = config.REL_PRODUCT_PATH + '/' + stem + '_group.dat',
                      minbincount = '20',
                      is_individual = False)
    
    # needs rel product path because absolute path is too long
    nutools.bin_pha(infile = config.REL_PRODUCT_PATH + '/' + stem + '.pha',
                    grpfile = config.REL_PRODUCT_PATH + '/' + stem + '_group.dat',
                    outfile = config.REL_PRODUCT_PATH + '/' + stem + '_bin.pha')
```

```python
# combine all spectra
for module in modules:
    input_obs = ['nu' + observation + module + region + '_sr.pha' 
                 for observation in observations]
    outfile = module + '_' + region
    nutools.addspec(config.EXPERIMENT_PATH, config.PRODUCT_PATH, input_obs, outfile)
```

```python
# group and bin fully combined spectra
for module in modules:
    stem = module + '_' + region
    
    nutools.group_pha(infile = config.REL_PRODUCT_PATH + '/' + stem + '.pha',
                      bgfile = config.REL_PRODUCT_PATH + '/' + stem + '.bak',
                      outfile = config.REL_PRODUCT_PATH + '/' + stem + '_group.dat',
                      minbincount = '20',
                      is_individual = False)
    
    # needs rel product path because absolute path is too long
    nutools.bin_pha(infile = config.REL_PRODUCT_PATH + '/' + stem + '.pha',
                    grpfile = config.REL_PRODUCT_PATH + '/' + stem + '_group.dat',
                    outfile = config.REL_PRODUCT_PATH + '/' + stem + '_bin.pha')
```
